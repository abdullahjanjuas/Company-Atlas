```python
import pandas as pd
```

```python
orgs = pd.read_csv("/kaggle/input/organizations-csv/organizations_with_hierarchical_clusters.csv")
clusters = pd.read_csv("/kaggle/input/organizations-csv/clusters.csv")
```

```python
app_df = orgs.merge(
    clusters,
    how="left",
    left_on="fine_cluster",
    right_on="fine_id"
)
```

```python
app_df.head(10)
```

```python
app_df["fine_label"] = app_df["fine_label"].fillna("Unclustered")
app_df["coarse_label"] = app_df["coarse_label"].fillna("Unclustered")
```

```python
app_df.head()
```

```python
app_df = app_df.drop(columns=["fine_id", "coarse_id", "fine_cluster", "coarse_cluster"])
```

```python
app_df.head()
```

```python
keep_cols = [
    "uuid", "name", "permalink", "cb_url",
    "short_description", "description",
    "fine_label", "coarse_label",
    "country_code", "region", "city", "status", "founded_on", "employee_count",
    "num_funding_rounds", "total_funding_usd",
    "homepage_url", "linkedin_url", "twitter_url"
]

app_df = app_df[keep_cols]
```

```python
app_df["founded_on"] = pd.to_datetime(
    app_df["founded_on"].astype(str).str[:10],  # cast to string first
    format="%Y-%m-%d",
    errors="coerce"
)

app_df["total_funding_usd"] = pd.to_numeric(app_df["total_funding_usd"], errors="coerce")
app_df["employee_count"] = pd.to_numeric(app_df["employee_count"], errors="coerce")
```

```python
app_df["founded_year"] = app_df["founded_on"].dt.year
```

```python
app_df.head()
```

```python
app_df["founded_year"] = app_df["founded_year"].astype("Int64")  # keeps NaN support
```

```python
app_df.to_csv("final_dataset.csv", index=False)
```
